```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
reward = []
with open('reward_list_fw1.txt', 'r') as f:
    for line in f:
        reward.append(float(line))

step = []
with open('step_list_fw1.txt', 'r') as f:
    for line in f:
        step.append(int(line))
```

```python
plt.plot(reward)
```

```python
avg_reward = []
for i in range(len(reward)):
    if i < 100:
        avg_reward.append(0)
    else:
        avg_reward.append(np.mean(reward[i-100:i]))
plt.plot(range(100,len(avg_reward)),avg_reward[100:])
```

```python
plt.plot(step)
```

```python
avg_step = []
for i in range(len(step)):
    if i < 100:
        avg_step.append(0)
    else:
        avg_step.append(np.mean(step[i-100:i]))
plt.plot(range(100,len(avg_step)),avg_step[100:])
```

```python
plt.plot(step[:len(reward)], reward)
plt.xlabel('Step')
plt.ylabel('Reward')
```
